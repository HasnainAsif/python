```python
# Example of K-means on IRIS data using SNS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
# load the dataset
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'].unique()
```

```python
df.info()
```

```python
# Here we are using only 2 dimensions(features) from dataset
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.show()
```

```python
X = df.drop(columns=['species'])
y = df['species']

test_values = np.array([[5.0, 3.5, 1.5, 0.2],
                        [6.5, 3.0, 5.2, 2.0],
                        [4, 2.0, 3.7, 6.4],
                        [4, 2.0, 3.7, 6.4]
                        ])
```

```python
# Create a KMeans instance with 3 clusters - Choosing three because there are three species of iris in the dataset
model = KMeans(n_clusters=3)
model.fit(X) # Fit model to dataframe

labels = model.predict(test_values) # Determine the cluster labels of new_points
print(labels) # Print cluster labels of new_points
```

```python
# Get coordinates of the centroids
centroids = model.cluster_centers_
print(centroids) # returned 3 rows because no. of clusters are 3

# Get columns of centroids
centroids_a = centroids[:,0]
centroids_b = centroids[:,1]
# centroids_c = centroids[:,2]
# centroids_d = centroids[:,3]

# Make a scatter plot of centroids_a and centroids_a - Took only two centroids because we can only plot in 2D
plt.scatter(centroids_a, centroids_b, marker='D', s=50)
plt.show()
```

### Find best suited `no. of cluster` for our model

```python
ks = range(1, 7)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k) # Create a KMeans instance with k clusters
    model.fit(X) # Fit model to samples
    inertias.append(model.inertia_) # Append the inertia to the list of inertias
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

Inertia decreases very slowly from 3 cluster to 4.

Hence, after 3 clusters there is almost no benefit for increasing clusters on KMeans

## Check clusters of Predicted and Actual labels using crosstab

```python
model = KMeans(n_clusters=3)

labels = model.fit_predict(X) # Use fit_predict to fit model and obtain cluster labels
df = pd.DataFrame({'labels': labels, 'varieties': y}) # Create a DataFrame with labels and varieties as columns

ct = pd.crosstab(df['labels'], df['varieties']) # Create crosstab
print(ct)
```

## Standard Scaler and predict using pipeline

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
model = KMeans(n_clusters=3)
pipeline = make_pipeline(scaler, model)
labels = pipeline.fit_predict(X) # Use fit_predict to fit model and obtain cluster labels

df = pd.DataFrame({'labels': labels, 'varieties': y}) # Create a DataFrame with labels and varieties as columns
ct = pd.crosstab(df['labels'], df['varieties']) # Create crosstab
print(ct)
```
