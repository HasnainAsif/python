```python
# import libraries
import numpy as np
import cv2 as cv
```

```python
# read the image
img = cv.imread("./images/test_01.jpg")
```

```python
# show image in separate window
cv.imshow("image", img)
cv.waitKey(0) # wait for a key press
cv.destroyAllWindows() # close the window after a key press
```

```python
# check the size of img
img.shape
```

```python
# convert the img into grascale and then check the shape
# convert the img into grayscale
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# check the shape of gray_img
gray_img.shape
```

```python
# read webcam using cv
cap = cv.VideoCapture(0) # 0 for default webcam, 1 for external webcam
if cap.isOpened():
    ret, frame = cap.read() # read the frame from webcam
else:
    ret = False # ret is True if the frame is read successfully
while ret:
    ret, frame = cap.read()
    cv.imshow("webcam", frame)
    if cv.waitKey(25) & 0xFF ==ord("q"): # press 'q' to exit. 25 is the delay in milliseconds
        break
cap.release() # release the webcam means it is no longer in use
cv.destroyAllWindows() # close the window after q is pressed
```

```python
# read webcam using cv
cap = cv.VideoCapture(0)

# change the cam into grayscale cam
while True:
    ret, frame = cap.read()
    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    # hls_frame = cv.cvtColor(frame, cv.COLOR_BGR2HLS)  # Convert to HLS color space
    
    cv.imshow("webcam", gray_frame)
    if cv.waitKey(25) & 0xFF == ord("q"):
        break

cap.release()
cv.destroyAllWindows()
```
