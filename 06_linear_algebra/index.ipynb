# Linear Algebra in Python

#### System of equations:
2x + 3y = 8

5x + 7y = 20

#### Matrix form
Ax = B

A = [[2, 3], [5, 7]]

B = [8, 20]



```python
import numpy as np

# coefficient matrix A
A = np.array([[2, 3], [5, 7]]) # 2 x 2 matrix
print('A:')
print(A)

# constant matrix B
B = np.array([8, 20]) # 2 x 1 matrix
print('B:')
print(B)

# Solve for x
x = np.linalg.solve(A, B)
print('x:')
print(x)
```

## Solving Complex Equations

\[
\begin{align*}
(1 + 2i)x + (2 - i)y &= 4 + i \\
(3 - i)x + 4y &= 3 + 4i
\end{align*}
\]

```python
# coefficients matrix A with complez numbers
A_complex = np.array([[1+2j, 2-1j], [3-1j, 4]]) # 2 x 2 matrix with complex numbers

# Constant matrix B with complex numbers
B_complex = np.array([4+1j, 3+4j]) # 2 x 1 matrix with complex numbers

# Solve for x with complex numbers
x_complex = np.linalg.solve(A_complex, B_complex)
print('x with complex numbers:')
print(x_complex)

# Verify the solution
verification = np.allclose(np.dot(A_complex, x_complex), B_complex)
print('Verification of complex solution:', verification)
```

## Draw Vectors

```python
import numpy as np
import matplotlib.pyplot as plt

# create a vector
v = np.array([2, 3]) # (2, 3) these are coordinates of v vector
print(v)

# plot the v vector
# plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(0, 5) # x limit i.e; x ranges from 0 to 5
plt.ylim(0, 5) # y limit i.e; y ranges from 0 to 5
plt.grid()
plt.title('2D Vector Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.gca().set_aspect('equal', adjustable='box') # set aspect ratio to equal i.e; it will make box of graph more fit to content
plt.show()
```

## Linear Transformation

### Simple Transformation

```python
import numpy as np
import matplotlib.pyplot as plt

# create a vector
v = np.array([2, 3]) # (2, 3) these are coordinates of v vector
print('Original vector:')
print(v)

# Rotation matrix for 45 degrees
# theta = np.pi / 4  # 45 degrees in radians
theta = np.radians(45)  # 45 degrees in radians
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], 
                            [np.sin(theta), np.cos(theta)]])

print('Rotation matrix:')
print(rotation_matrix)

# Apply transformation/rotation to vector v
v_transformed = np.dot(rotation_matrix, v) # multiplying Original vector with transformation matrix, will return transformed vector
print('Transformed/Rotated vector:')
print(v_transformed)

# Plot the original and transformed/rotated vectors
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector')
plt.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='b', label='Transformed Vector')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.title('2D Vector Transformation - by 45 Degrees')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

### Shear Transformation

```python
import numpy as np
import matplotlib.pyplot as plt

# create a vector
v = np.array([2, 3]) # (2, 3) these are coordinates of v vector
print('Original vector:')
print(v)

# apply shear transformation on original vector
# Shear matrix

shear_matrix = np.array([[1, 1.5], [0, 1]])  # Shear along y-axis

print('Shear matrix:')
print(shear_matrix)

# Apply shear transformation to vector v
v_sheared = np.dot(shear_matrix, v)
print('Sheared vector:')
print(v_sheared)

# Plot the original and sheared vectors
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector')
plt.quiver(0, 0, v_sheared[0], v_sheared[1], angles='xy', scale_units='xy', scale=1, color='b', label='Sheared Vector')
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.grid()
plt.title('2D Vector Shear Transformation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()

######## Adjusting the plot to include the effect of shearing on the grid lines

# Change grid lines to reflect shear transformation
grid_x, grid_y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
grid_points = np.vstack([grid_x.ravel(), grid_y.ravel()])
sheared_grid_points = np.dot(shear_matrix, grid_points)

# Plot the sheared grid
plt.figure()
for i in range(10):
    plt.plot(sheared_grid_points[0, i*10:(i+1)*10], sheared_grid_points[1, i*10:(i+1)*10], color='gray', linestyle='--', linewidth=0.5)
    plt.plot(sheared_grid_points[0, i::10], sheared_grid_points[1, i::10], color='gray', linestyle='--', linewidth=0.5)

# Re-plot the original and sheared vectors for context
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector')
plt.quiver(0, 0, v_sheared[0], v_sheared[1], angles='xy', scale_units='xy', scale=1, color='b', label='Sheared Vector')
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.grid()
plt.title('2D Vector Shear Transformation with Sheared Grid')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

### SVD In Python

```python
# generate a matrix of random numbers of 4x3 dataset
A = np.random.rand(4, 3)
print('A:')
print(A)

# lets apply SVD here
U, Sigma, VT = np.linalg.svd(A)
print("U Matrix: ")
print(U)
print("Sigma Matrix:")
print(Sigma)
print('V transpose Matrix:')
print(VT)
```
