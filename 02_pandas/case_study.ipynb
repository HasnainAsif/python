```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# read titanic dataset
kashti = sns.load_dataset("titanic")
print(kashti.head())
```

```python
# drop columns
kashti.drop(columns=["deck", "alone"]).head()
```

```python
# check how many survived and how many did not
# kashti["survived"].value_counts()
kashti.value_counts("survived")
```

```python
kashti.info()
```

```python
# kashti.groupby(['sex'], observed=True).mean(numeric_only=True)
print(kashti.groupby(['sex', 'class'], observed=True).mean(numeric_only=True))
```

```python
# children survival rate
kashti[kashti['age'] < 18].mean(numeric_only=True)
```

```python
# grouped children survival rate
kashti[kashti['age'] < 18].groupby(['sex', 'class']).mean(numeric_only=True)
```

```python
# number of children and their survival rate
kashti[kashti['age'] < 18].groupby(['sex', 'class'], observed=True)["survived"].agg(["mean", "count"])
```
