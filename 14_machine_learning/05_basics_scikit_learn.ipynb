# Scikit-Learn Library Basic Usage

## Linear Regression Model

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
# load the data
df = sns.load_dataset('tips')
df.head()
```

```python
# Create "regression line" plot
sns.lmplot(data=df, x="total_bill", y="tip") # regplot can also be used
## Line in graph is best fit line
```

```python
# Select the feature column from df
X = df['total_bill']
# Select the target/label column from df
Y = df['tip']
```

```python
# load the linear regression model
model = LinearRegression()

# Reshape X to be 2D as required by sklearn
X_reshaped = X.values.reshape(-1, 1) # X = df[['total_bill']] --> return df of total_bill instead of series, if we dont want to reshape "X".
# print(X_reshaped)

# Fit the model --> means; train the model
model.fit(X_reshaped, Y)
```

```python
# lets predict the tip for an unknown total_bill
model.predict([[50]]) # let total_bill = 50. find what will be tip?
```
