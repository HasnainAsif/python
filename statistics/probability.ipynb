## General Probability Formula
$P(E) = \dfrac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}$

---

## üé≤ Dependent and Independent Events in Probability  

**1. Independent Events**  
- Two events are **independent** if the outcome of one **does not affect** the outcome of the other.  
- Example: Tossing a coin and rolling a dice.  
  - Probability of Head = 1/2  
  - Probability of rolling a 4 = 1/6  
  - Probability of both happening = (1/2) √ó (1/6) = 1/12  

**Formula:**
$P(A \cap B) = P(A) \times P(B)$

---

**2. Dependent Events**  
- Two events are **dependent** if the outcome of one **does affect** the outcome of the other.  
- Example: Drawing cards from a deck **without replacement**.  
  - Probability of first card being Ace = 4/52  
  - Now only 51 cards left ‚Üí Probability of second Ace = 3/51  
  - Probability of both Aces = (4/52) √ó (3/51)  

**Formula:**  
$P(A \cap B) = P(A) \times P(B|A)$

(where \(P(B|A)\) = probability of B happening after A has already happened)

---

## ‚ôªÔ∏è With Replacement vs Without Replacement  

**üëâ With Replacement**  
- After drawing/choosing, the item is **put back**.  
- The total number of outcomes **stays the same**.  
- Example: Drawing a card, putting it back, then drawing again.  
  - Probability of Ace in both draws = (4/52) √ó (4/52)  

**üëâ Without Replacement**  
- After drawing/choosing, the item is **not put back**.  
- The total number of outcomes **reduces**.  
- Example: Drawing two cards without putting the first back.  
  - Probability of two Aces = (4/52) √ó (3/51)  

---

‚úÖ **Summary**  
- **Independent events:** Outcomes don‚Äôt affect each other.  
- **Dependent events:** Outcomes affect each other.  
- **With Replacement:** Same probability each time (independent).  
- **Without Replacement:** Probability changes after each draw (dependent).  


# Calculate Probability

```python
import pandas as pd

# import csv
amir_deals = pd.read_csv('../00_datasets/amir_deals.csv')

counts = amir_deals['product'].value_counts() # Count the deals for each product
print(counts)
total_deals = counts.sum() # Total number of deals
# total_deals = amir_deals['product'].count() # Total number of deals (alternative way)
print(total_deals)

# Calculate probability of picking a deal with each product
probabilities = counts / total_deals
print('-------------------------------')
print(probabilities)\

# If you randomly select one of Amir's deals, what's the probability that the deal will involve Product C?
print('-------------------------------') 
prob_c = probabilities['Product C'] # Access the probability for Product C
print(str(prob_c * 100) + " %")  
```

# Discrete distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create DataFrame
restaurant_groups = pd.DataFrame({
    "group_id": list("ABCDEFGHIJ"),
    "group_size": [4,2,4,3,2,2,6,2,2,2]
})
# print(restaurant_groups)

# # Create a histogram of restaurant_groups and show plot
restaurant_groups['group_size'].hist(bins=[2, 3, 4, 5, 6])
plt.show()

# Create probability distribution
size_dist = restaurant_groups['group_size'].value_counts() / len(restaurant_groups) # Calculate the probability of each group size
# print(size_dist)
size_dist = size_dist.reset_index()
# print(size_dist)
size_dist.columns = ['group_size', 'prob'] # Rename columns to group_size and prob
print(size_dist)

# Calculate expected value
expected_value = np.sum(size_dist['group_size'] * size_dist['prob']) # Expected value is the sum of each group size multiplied by its probability
# expected values means the average group size
print(expected_value)

# Subset groups of size 4 or more
groups_4_or_more = restaurant_groups[restaurant_groups['group_size'] >= 4]
# print(groups_4_or_more)

# Sum the probabilities of groups_4_or_more
prob_4_or_more = groups_4_or_more['group_size'].value_counts() / len(restaurant_groups)
print(prob_4_or_more.sum())
```

# Continuous Uniform Distribution

```python
# The sales software used at your company is set to automatically back itself up. back-ups happen exactly every 30 minutes.
# He wants to know how long he'll have to wait if he adds some data, to get backed up.
# Min and max wait times for back-up that happens every 30 min

from scipy.stats import uniform
import matplotlib.pyplot as plt

min_time = 0
max_time = 30


# Calculate probability of waiting less than 5 mins
prob_less_than_5 = uniform.cdf(5, loc=min_time, scale=max_time - min_time)  # CDF gives the probability of waiting less than a certain time
print(prob_less_than_5)

# Calculate probability of waiting more than 5 mins
prob_greater_than_5 = uniform.cdf(30, loc=min_time, scale=max_time - min_time) - prob_less_than_5
print(prob_greater_than_5)

# Calculate probability of waiting 10-20 mins
prob_between_10_and_20 = uniform.cdf(20, loc=min_time, scale=max_time - min_time) - uniform.cdf(10, loc=min_time, scale=max_time - min_time)
print(prob_between_10_and_20)

# Generate 1000 wait times between 0 and 30 mins
wait_times = uniform.rvs(min_time, max_time, 1000)  # Generate random wait times
print(wait_times)

# Create a histogram of simulated times and show plot
plt.hist(wait_times)
plt.show()
```

# Binomial Distribution

```python
# Below example, mentions that a sales person has a 30% chance of winning a deal.

# Import binom from scipy.stats
from scipy.stats import binom
import numpy as np

# Set random seed to 10
np.random.seed(10)

# Simulate a single deal - Simulate a single deal with a 30% chance of success 
deal = binom.rvs(n=1, p=0.3, size=1) # rvs generates random variates from a binomial distribution
# n is the number of trials (1 deal), p is the probability of success (0.3), size is the number of simulations (1)
print(deal)

# Simulate 1 week of 3 deals
deals = binom.rvs(n=3, p=0.3, size=1) # Simulate 1 week of 3 deals with a 30% chance of success
print(deals)

# Simulate 52 weeks of 3 deals
deals_new = binom.rvs(n=3, p=0.3, size=52) # Simulate 52 weeks of 3 deals with a 30% chance of success
print(deals_new)
mean_deals_won = deals_new.mean()  # Calculate the mean number of deals won per week
print(mean_deals_won) # Print mean deals won per week

# Probability of closing 3 out of 3 deals
prob_3_out_of_3 = binom.pmf(k=3, n=3, p=0.3)  # PMF gives the probability of exactly k successes in n trials
print(prob_3_out_of_3)

# Probability of closing <= 1 deal out of 3 deals
prob_less_than_equal_1 = binom.cdf(k=1, n=3, p=0.3)  # CDF gives the probability of k or fewer successes in n trials
print(prob_less_than_equal_1)

# Probability of closing > 1 deal out of 3 deals
prob_greater_than_1 = 1 - prob_less_than_equal_1  # Probability of more than 1 deal is 1 minus the probability of 1 or fewer deals
print(prob_greater_than_1)

```

# expected value of a binomial distribution 

- It can be calculated by n * p.

```python
# Calculate the expected number of sales out of the 3 he works on that Amir will win each week if he maintains his 30% win rate
won_30pct = 3 * 0.3 # Expected number won with 30% win rate
print(won_30pct)

# Expected number won with 25% win rate
won_25pct = 3 * 0.25
print(won_25pct)

# Expected number won with 35% win rate
won_35pct = 3 * 0.35
print(won_35pct)
```

# Normal Distribution of Probability distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# load csv
amir_deals = pd.read_csv('../00_datasets/amir_deals.csv')
amir_deals.head()

# Histogram of amount with 10 bins 
amir_deals['amount'].hist(bins=10)
plt.show()

# Probability of deal < 7500
prob_less_7500 = norm.cdf(7500, 5000, 2000) # 5000 is mean, 2000 is std deviation
print(prob_less_7500)

# Probability of deal > 1000
prob_greater_1000 = 1 - norm.cdf(1000, 5000, 2000)
print(prob_greater_1000)

# Probability of deal between 3000 and 7000
prob_3000_to_7000 = norm.cdf(7000, 5000, 2000) - norm.cdf(3000, 5000, 2000)
print(prob_3000_to_7000)

# Calculate amount that 25% of deals will be less than
pct_25 = norm.ppf(0.25, 5000, 2000)
print(pct_25)

```
