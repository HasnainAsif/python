```python
import wbdata
import pandas as pd

# define indicator for population (SP.POP.TOTL)
indicator = {'SP.POP.TOTL': 'total_population', # population
             'SM.POP.NETM': 'net_migration' # net migration
             }
# define countries of interest
countries = ['IN', 'PK', 'BD', 'LK', 'AF'] # India, Pakistan, Bangladesh, Sri Lanka, Afghanistan

# fetch data
data = wbdata.get_dataframe(indicator, country=countries)

# reset index to make 'country' and 'date' columns
data.reset_index(inplace=True)

# rename columns for clarity
data.rename(columns={'country': 'Country', 'date': 'Year'}, inplace=True)

# convert 'Year' to numeric
data['Year'] = pd.to_numeric(data['Year'])

# filter data between 1960 and 2023
data = data[(data['Year'] >= 1960) & (data['Year'] <= 2023)]

# save data to csv file
data.to_csv('./datasets/population_migration.csv')

print(data.head())
```
