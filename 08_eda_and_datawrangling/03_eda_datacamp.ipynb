```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import csv
unemployment = pd.read_csv('../00_datasets/clean_unemployment.csv')
print(unemployment.head())
```

```python
plt.clf()
# Create a histogram of 2021 unemployment; show a full percent in each bin
sns.histplot(data=unemployment, x='2021', binwidth=1)
plt.show()
```

`Graph shows in 2021, most countries unemployment hovered around 3% to 9%`

`A very few countries, contain very low unemployment rate`

`A few countries, contain too high unemployment rate`


```python
# Check distribution of plot using kde
sns.kdeplot(data=unemployment, x="2021", 
            hue='continent'
            # cut=0, # minimum value should be greater than 0
            # cumulative=True
            )
plt.show()
```

```python
# we have been informed that the data on unemployment from countries in Oceania is not reliable. So, we need to filter this continent.

# not_Oceania = unemployment['continent'] != 'Oceania'
not_oceania = ~unemployment['continent'].isin(['Oceania']) # continents not equal to Oceania
unemployment = unemployment[not_oceania]
print(unemployment['continent'].value_counts())
print(unemployment.shape)
```

```python
# Print the minimum and maximum unemployment rates during 2021
print(unemployment['2021'].min(), unemployment['2021'].max())

# Create a boxplot of 2021 unemployment rates, broken down by continent
sns.boxplot(data=unemployment, x='continent', y='2021')
plt.xticks(rotation=45)
plt.show()
```

```python
# Print the mean and standard deviation of rates for 2019 and 2020 
print(unemployment[["2019", "2020"]].agg(['mean', 'std']))

print('----------------------------------------------')

# Print mean and standard deviation grouped by continent
print(unemployment[["continent", "2019", "2020"]].groupby('continent').agg(['mean', 'std']))
```

```python
continent_summary = unemployment.groupby("continent").agg(
    # Create the mean_rate_2021 column
    mean_rate_2021=('2021', 'mean'),
    # Create the std_rate_2021 column
    std_rate_2021=('2021', 'std')
)
print(continent_summary)
```

```python
# Create a bar plot of continents and their average unemployment in 2021
sns.barplot(data=unemployment, x='continent', y='2021') # black lines are representing 95% confidence interval
plt.show()
```

`While Europe has higher average unemployment than Asia, it also has a smaller confidence interval for that average, so the average value is more reliable for Europe.`
