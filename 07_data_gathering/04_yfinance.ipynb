# load stock market data using yfinance library

```python
import pandas as pd
import yfinance as yf
```

```python
# define ticker
ticker = 'NVDA'
```

```python
# define the time period
start_date = '2015-01-01'
end_date = '2024-11-21'
```

```python
# download the data from yfinance
df_nvda = yf.download(ticker, start=start_date, end=end_date)
df_nvda.head()
```

```python
# Define each column in the dataset
# Date --> The date of the stock price
# Open --> The price of the stock at market open
# High --> The highest price of the stock during the trading day
# Low --> The lowest price of the stock during the trading day
# Close --> The price of the stock at market close
# Adj Close --> The adjusted closing price of the stock, accounting for dividends and stock splits
# Volume --> The number of shares traded during the day
```

```python
# Save the DataFrame to a CSV file
df_nvda.to_csv('./datasets/'+ ticker +'_yfinance.csv')
```

```python
# download data from multiple tickers
tickers = ['NVDA', 'AAPL', 'MSFT', 'GOOGL', 'AMZN'] 
df_multiple = yf.download(tickers, start="2010-01-01", end="2024-12-31", auto_adjust=True)

df_multiple.head()

# save the DataFrame to a CSV file
df_multiple.to_csv('./datasets/'+ '_'.join(tickers) +'_yfinance.csv')
```
