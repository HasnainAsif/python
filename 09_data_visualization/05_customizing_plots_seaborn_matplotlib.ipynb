# Customizing Plots in Seaborn and Matplotlib

## Loading Libraries and Data

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
# Load the sample diamonds data
df = pd.read_csv('../00_datasets/diamonds.csv')
print(f"Number of rows before sampling: {df.shape[0]}")
print(f"Number of columns before sampling: {df.shape[1]}")
print(f"{50*'-'}")
df = df.sample(5000) # Sampling 5000 rows from the dataset of 53940 rows
print(f"Number of rows after sampling: {df.shape[0]}")
print(f"Number of columns after sampling: {df.shape[1]}")

# Display the first few rows of the dataset
df.head()
```

## 1. Setting Titles

```python
sns.scatterplot(x='carat', y='price', data=df)
```

```python

sns.scatterplot(x='carat', y='price', data=df)
plt.title('Scatter Plot of Carat vs Price')  # Setting the title
plt.show()
```

## 2. Setting Axes Labels

```python

sns.scatterplot(x='carat', y='price', data=df)
plt.xlabel('Carat of Diamond')  # Label for x-axis
plt.ylabel('Price ($)')  # Label for y-axis
plt.show()
```

## 3. Customizing the Legend

```python
sns.scatterplot(x='carat', y='price', hue='cut', data=df)
```

```python

sns.scatterplot(x='carat', y='price', hue='cut', data=df)
plt.legend(title='Diamond Cut', loc='lower right')  # Customizing the legend
plt.show()
```

## 4. Customizing Ticks

```python

sns.histplot(df['price'], bins=10, edgecolor='blue')
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)  # Change the font size of y-axis labels
plt.show()
```

## 5. Adding a Grid

```python

# sns.set_style('whitegrid')  # Use a grid background
sns.scatterplot(x='carat', y='price', data=df)
plt.grid(True, linestyle='--', alpha=0.5)  # Customize grid lines
plt.show()
```

## 6. Customizing Spines

```python

sns.set_theme(style='whitegrid')  # Use a white grid background
sns.boxplot(x='cut', y='price', data=df)
sns.despine()  # Remove the top and right spines for a cleaner look
plt.show()
```

## 7. Setting Figure Size

```python

plt.figure(figsize=(10, 6))  # Set figure size - 10 inches wide and 6 inches tall
sns.barplot(x='cut', y='price', data=df)
plt.title('Bar Plot of Price by Cut (Large Figure)')
plt.show()
```

## 8. Creating Subplots

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # Create a figure with 1 row and 2 columns of subplots
sns.histplot(df['carat'], bins=10, edgecolor='black', ax=axes[0], kde=True)
axes[0].set_title('Histogram of Carat')
sns.histplot(df['price'], bins=10, edgecolor='black', ax=axes[1], kde=True)
axes[1].set_title('Histogram of Price')
plt.show()
```

## 9. Adding Annotations

```python

df.reset_index(drop=True, inplace=True)  # Reset the index of the DataFrame

sns.scatterplot(x='carat', y='price', data=df)
plt.title('Scatter Plot with Annotations')
for i in range(len(df)):
    plt.annotate(f"${df['price'][i]}", (df['carat'][i], df['price'][i]), fontsize=8, alpha=0.7)
plt.show()
```

```python
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Scatter Plot with Annotations')
plt.annotate(f"${df['price'][10]}", 
             (df['carat'][10], df['price'][10]), # Coordinates for the annotation
             xytext=(1.5, 100),  # Offset for the annotation text
             arrowprops=dict(arrowstyle='->', color='red'),  # Arrow properties
             fontsize=8, alpha=0.7)
plt.show()
```

## 10. Customizing Color Palettes

```python

figure = plt.figure(figsize=(15, 4))
sns.set_palette('muted') # muted, deep, pastel, bright, dark, colorblind
sns.boxplot(x='cut', y='price', data=df, hue='clarity')
plt.title('Box Plot with Muted Color Palette')
plt.show()
```
